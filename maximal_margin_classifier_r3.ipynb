```python

import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python

x1 = np.random.normal(30, 200, 150)

y1 =  np.random.normal(50, 300, 150)

z1 = np.random.normal(10, 10, 150)


y2 = np.random.normal(2000, 400, 150)

x2 =  np.random.normal(1000, 400, 150)

z2 = np.random.normal(900, 10, 150)
```

```python
# candidato ao maximal margin classifier

p = np.array([7, 4, 450]) #ponto no hyperplano

beta = np.array([0,1,2]) #vetor normal ao hyperplano

beta0 = np.inner(p, beta)


#modulo_beta = np.sqrt(beta[0]**2 + beta[1]**2 + beta[2]**2) #tamanho de beta

#margem = 1/modulo_beta


x = np.linspace(-1500 , 1500 , 100)
y = np.linspace(-1500 , 1500 , 100)

a, b = np.meshgrid(x, y)

Hyperplane = (beta0 -(a*beta[0])-(b*beta[1]))/beta[2]


fig = plt.figure(dpi=200)


ax = plt.axes(projection ='3d')



ax.scatter(x1, y1, z1, c = 'blue', label = 'Classe -1')
ax.scatter(x2, y2, z2, c = 'red', label = 'Classe +1')

ax.plot_wireframe(a, b, Hyperplane, alpha = 0.5, color = "grey")



ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.legend()

ax.view_init(20, 140)

plt.savefig("C:/Users/Igor/Desktop/TÓPICOS 2/max_margin_classifier_R3.pdf", format="pdf", bbox_inches="tight", dpi = 100)
plt.show()
```
